```python
yellow_tip.shape
green_tip.shape
```

```python
import pandas as pd
import numpy as np

# Independent T-Test
from scipy import stats
t_stat,p_val=stats.ttest_ind(yellow_tip,green_tip)
print(t_stat,p_val)

if p_val < 0.05:
    print("Reject null hypothesis: The tip for yellow category is different that of green")
else:
    print("Fail to reject null hypothesis")
```

Null Hypothesis(H0) The tip for Credit Card and Cash are same

Alternative Hypothesis(H1): The tip for Credit card and cash are different

```python
credit_tip=df['tip'][df['payment']=='credit card']
cash_tip=df['tip'][df['payment']=='cash']

# Independent T-test
t_stat,p_val=stats.ttest_ind(credit_tip,cash_tip)
print(t_stat,p_val)

if p_val<0.05:
  print("Reject null hypothesis: The tip for credit card is different that of cash")
else:
  print("Fail to reject null hypothesis")
```

**Hypothesis**:

Null Hypothesis(H0) The tip same across all the pickup_boroughs

Alternative Hypothesis(H1): The tip distribution is different for atleast one pickup_boroughs

```python
manhattan_tip=df['tip'][df['pickup_borough']=='Manhattan']
queens_tip=df['tip'][df['pickup_borough']=='Queens']
bronx_tip=df['tip'][df['pickup_borough']=='Bronx']
brooklyn_tip=df['tip'][df['pickup_borough']=='Brooklyn']
```

```python
# ANOVA Test

f_stat,p_val=stats.f_oneway(manhattan_tip,queens_tip,bronx_tip,brooklyn_tip)
print(f_stat,p_val)

if(p_val<0.05):
  print("Reject null hypothesis: The tip distributions are different for atleast one pickup_borough")
else:
  print("Fail to reject null hypothesis")
```

```python
# Tukey HSD Test
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Prepare data

# Perform Tukey HSD
tukey_result = pairwise_tukeyhsd(endog=df['tip'], groups=df['pickup_borough'], alpha=0.05)
print(tukey_result) # Gives the hypothesis result(to reject null or accept) for each possible pair of groups in reject column
```

**Correlation Hypothesis**:

Null Hypothesis(H0) There is no correlation

Alternative Hypothesis(H1): There is relation b/w 2 cols


```python
r,p_val=stats.pearsonr(df.passengers,df.tip)
print(r,p_val)
alpha=0.05
if p_val<alpha:
  print("Reject null hypothesis: There is a relationship between passengers and tip")
  if r>0:
    print("Positive correlation")
  else:
    print("Negative correlation")
else:
  print("Fail to reject null hypothesis")
```
